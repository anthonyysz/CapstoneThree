## 3. Modeling

### 3a. ARIMA Model

I'm going to define a couple of functions, and use them to find the most optimal ARIMA model perameters. 

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
fig,ax = plt.subplots(3,2, figsize = (10,10))

a = ax[0,0].plot(df['value']); ax[0,0].set_title('Original Series')
b = plot_acf(df['value'].values,ax=ax[0,1])

c = ax[1,0].plot(df['value'].diff()); ax[1,0].set_title('1st Difference')
d = plot_acf(df['value'].diff().dropna(),ax=ax[1,1])

e = ax[2,0].plot(df['value'].diff().diff()); ax[2,0].set_title('2nd Difference')
f = plot_acf(df['value'].diff().diff().dropna(),ax=ax[2,1])
```

Right now I am in between 1 and 2 for my d value. I fear that 1 might be under differencing the data, and 2 might be over differencing it. I am going to plot the pacf for both so that I can make a better decision.

```python
import warnings
from statsmodels.tsa.arima.model import ARIMA
warnings.filterwarnings('ignore')

order_aic_bic=[]

# Loop over p values from 0-2
for p in range(3):
  # Loop over q values from 0-2
    for q in range(3):
      	# create and fit ARMA(p,q) model
        model = ARIMA(df['value'], order=(p, 1, q))
        results = model.fit()
        
        # Append order and results tuple
        order_aic_bic.append((p, q,results.aic, results.bic ))
```

```python
order_df = pd.DataFrame(order_aic_bic, 
                        columns=['p', 'q', 'AIC', 'BIC'])

print(order_df.sort_values('AIC'))
print(order_df.sort_values('BIC'))
```

```python
order_aic_bic=[]

# Loop over p values from 0-2
for p in range(3):
  # Loop over q values from 0-2
    for q in range(3):
      	# create and fit ARMA(p,q) model
        model = ARIMA(df['value'], order=(p, 2, q))
        results = model.fit()
        
        # Append order and results tuple
        order_aic_bic.append((p, q,results.aic, results.bic ))
```

```python
order_df = pd.DataFrame(order_aic_bic, 
                        columns=['p', 'q', 'AIC', 'BIC'])

print(order_df.sort_values('AIC'))
print(order_df.sort_values('BIC'))
```

We got much lower AIC and BIC values using a d value of 2, so our model will be created using the order (1, 2, 2).

Here, since the Ljung-Box p value is above .05, the null hypothesis is not rejected and we cannot say for sure that the residuals are uncorrelated. Additionally, however, we also reject the Heteroskedasticity null hypothesis, which means that the residuals in the forecast don't show variance over time. 

```python
y_pred = model1fit.get_forecast(len(test.index))
y_pred_df = y_pred.conf_int(alpha = 0.05) 
y_pred_df["Predictions"] = model1fit.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
y_pred_df.index = test.index
y_pred_out = y_pred_df["Predictions"] 
plt.plot(y_pred_out, color='blue', label = 'ARIMA Predictions')
plt.plot(train['value'], color = "black", label = 'Train Data')
plt.plot(test['value'], color = "red", label = 'Test Data')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

arma_rmse = np.sqrt(mean_squared_error(test["value"].values, y_pred_df["Predictions"]))
print("RMSE: ",arma_rmse)
```
